```python
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
def six_dice():
    r = np.ceil(np.random.uniform(0, 6))
    return r
```

```python
N = 1000
dice_results = np.array([six_dice() for n in range(N)])
print(dice_results.max(), dice_results.min())
# pprint(dice_results)
```

```python
bins = np.array([i + 1 for i in range(6 + 1)], dtype=int)
print(bins)
dice_histogram = np.histogram(dice_results, bins=bins)
print(dice_histogram)
```

```python
plt.hist(dice_results, bins=bins)
plt.show()
```

```python
def six_dices(D):
    dice_results = np.array([six_dice() for d in range(D)], dtype=int)
    return dice_results.sum()
```

```python
def bin_dices(D):
    return np.array([i + 1 for i in range(6 * D + 1)], dtype=int)
```

```python
def plot_dices_results(N, D):
    plt.figure(D)
    name = str(D) + '_dices'
    plt.title(name)
    dices_results = np.array([six_dices(D) for n in range(N)])
    bins = bin_dices(D)
    plt.hist(dices_results, bins=bins, density=True)
    plt.savefig('img/' + name + '.png')
#     plt.show()
    return dices_results, bins
```

```python
N = 100
D = 2
dices_results, bins = plot_dices_results(N, D)
```

```python
for D in np.arange(1, 20 + 1):
    N = D * 1000
    dices_results, bins = plot_dices_results(N, D)
    print(D, 'dices', N, 'throws')
    print('Average =', dices_results.mean(), 'std =', dices_results.std())
    print()
```
